#### What is MLflow Tracking?
- Model Metrics
    - F1, Recall, Accuracy, MSE, etc ...
- Parameters
    - library specific
- Code

- Other artifacts
    - tokenizers, pickle, etc...

#### Logging to MLflow Tracking
- Metrics
    - `log_metric("accuracy", 0.90)`
    - `log_metrics({"accuracy": 0.90, "loss": 0.50})`
- Parameters
    - `log_param("n_jobs":1 )`
    - `log_params({("n_jobs":1,"fit_intercept": False})`

- Artifacts
    - `log_artifact('file.py')`
    - `log_artifacts("./directory/")`

#### Querying Runs
- Searching runs

```python
mlflow.search_runs()
```

#### Filtering run searches
    - `max_results` - maximum number of results to return.
    - `order_by` - column(s) to sort in `ASC` ending or `DESC` ending order.
    - `filtering_string` - string based query.

#### Search runs example

```python
from mlflow.metrics import f1_score
import mlflow
# Filter string
f1_score_filter = "metrics.f1_score > 0.60"
# Search runs
mlflow.search_runs(experiment_names=["Insurance Experiment"], filter_string=f1_score_filter, order_by=["metrics.precision_score DESC"])
```
