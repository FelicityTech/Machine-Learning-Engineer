Introduction to MLflow Projects

MLflow Projects

- Reproducible
- Reusable
- Portable

MLproject file
- `name:` - Name of the Project
- `entry_points:`
    - Command(s) to run
    - `.py` and `.sh` files
    - Workflows
- `python_env:`
    - Python environment
    - `python_env:yaml`

MLproject example

```python
name: salary_model
entry_posts:
    main
    :
        command: "python train_model.py"
python_env: python_env.yaml
```

train_model.py

```python
# import libraries and modules
import mlflow
import mlflow.sklearn
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Training Data
df = pd.read_csv('Salary_predict.csv')
X = df[["experience", "age", "interview_score"]]
y = df[["Salary"]]
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=0)

```

```python
# Set Auto logging for Scikit-learn flavor
mlflow.sklearn.autolog()

# Train the model
lr = LinearRegression()
lr.fit(X_train, y_train)
```

Running MLflow Projects

### Projects API

`mlflow.projects`

`mlflow.projects.run()`
- `uri` - URI to MLprojct file
- `entry_point` - Entry point to start run from MLproject
- `experiment_name` - Experiment to track training run
- `env_manager` - Python environment manager: `local` or `virtualenv`

```python
# Run MLflow project
mlflow.projcts.run(
    uri='./',
    entry_point='main',
    experiment_name = 'My Experiment',
)
```

MLflow projects module

```python
import mlflow

# Set the run function from the MLflow Projects module
mlflow.projects.run(
  	# Set the URI as the current working directory
    uri='./',
    # Set the entry point to main
    entry_point='main',
    # Set the experiment name as Insurance
    experiment_name='Insurance',
    env_manager="local",
    synchronous=True,
)
```

Adding parameters to project run

```python
import mlflow

# Set the run function from the MLflow Projects module
mlflow.projects.run(
    uri='./',
    entry_point='main',
    experiment_name='Insurance',
  	env_manager='local',
  	# Set parameters for n_jobs and fit_intercept
  	parameters={
        'n_jobs_param': 2, 
    'fit_intercept_param': False
    }
)
```
