### Putting it all together!

```python
# Define extract(), transform(), and load() functions for an ETL pipeline

def extract(file_name, values):
    # Some custom-built Python to extract data from a source
    print(f"Extracting data from {file_name}")
    return pd.read_csv(file_name)

def transform(data_frame, values):
    return data_frame.loc[data_frame["name"] == values, ["name", "price"]]

# First, extract the data from csv
extract_data = extract(file_name="raw_data.csv", values="Amazon Echo")

# Then, transform the extracted data
transformed_data = transform(extract_data, values="Amazon Echo")

# Filally, load the transformed data to a target destination
load(data_frame=transformed_data, target_table="cleaned_data")
```

Building an ETL Pipeline

```python
def load(data_frame, file_name):
  # Write cleaned_data to a CSV using file_name
  data_frame.to_csv(file_name)
  print(f"Successfully loaded data to {file_name}")

extracted_data = extract(file_name="raw_data.csv")

# Transform extracted_data using transform() function
transformed_data = transform(data_frame=extracted_data)

# Load transformed_data to the file transformed_data.csv
load(data_frame=extracted_data, file_name="transformed_data.csv")
```
